# Rack temperature data

```python
rd = pd.read_csv('data/rack-temps.csv', parse_dates={'datetime' : ['Date time']}, date_parser=lambda x: datetime.strptime(x, '%d/%m/%Y %H:%M'))
dates = rd['datetime']
data  = rd.drop('datetime', 1).replace(',','.', regex=True).astype('float')

rd = data.set_index(pd.DatetimeIndex(dates))
rd.to_pickle('data/rack_temps.pickle')
rd.head()
```

```python
end = datetime(2014,3,31)
start = end - timedelta(days=7)

train = rd[:start-timedelta(seconds=1)]
train.to_pickle('data/rack_temps_train.pickle')

test = rd[start:end]
test.to_pickle('data/rack_temps_test.pickle')

#test.head()
train.head()
```
